```python
import pandas as pd
import matplotlib.pyplot as plt

# Nombres de los archivos CSV
archivos = ['resultado-tfm-rag.csv', 'resultados-tfm-rag-chunk-2000.csv', 'resultado-tfm-rag-preprocessed.csv', 
            'resultado-tfm-rag-preprocessed-2000.csv', 'resultados-tfm-rag-cohere.preprocesadoManual-chunk-2000.csv', 'resultados-tfm-rag-cohere-preprocesadoLLM-chunk-2000.csv']

# Nombre de las mtricas
metricas = ['context_precision', 'context_recall', 'faithfulness', 'answer_relevancy']

# Cargar los datos de cada archivo en un diccionario
escenarios = {}
for i, archivo in enumerate(archivos):
    escenario = f'Escenario {chr(65 + i)}'  # Escenario A, B, C, etc.
    escenarios[escenario] = pd.read_csv(archivo)

# Crear los grficos
for escenario, datos in escenarios.items():
    plt.figure(figsize=(10, 6))
    
    # Dibujar cada mtrica
    for metrica in metricas:
        plt.plot(range(1, 18), datos[metrica], marker='o', label=metrica)

    # Aadir detalles al grfico
    plt.title(f'Resultados de Evaluacin - {escenario}')
    plt.xlabel('Nmero de Pregunta')
    plt.ylabel('Valor de la Mtrica')
    plt.ylim(0, 1)  # Todas las mtricas estn en el rango 0 a 1
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Nombres de los archivos CSV
archivos = ['resultado-tfm-rag.csv', 'resultados-tfm-rag-chunk-2000.csv', 'resultado-tfm-rag-preprocessed.csv', 
            'resultado-tfm-rag-preprocessed-2000.csv', 'resultados-tfm-rag-cohere.preprocesadoManual-chunk-2000.csv', 'resultados-tfm-rag-cohere-preprocesadoLLM-chunk-2000.csv']

# Nombre de las mtricas
metricas = ['context_precision', 'context_recall', 'faithfulness', 'answer_relevancy']

# Cargar los datos de cada archivo en un diccionario
escenarios = {}
for i, archivo in enumerate(archivos):
    escenario = f'Escenario {chr(65 + i)}'  # Escenario A, B, C, etc.
    escenarios[escenario] = pd.read_csv(archivo)

# Crear los colores para cada escenario
colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']  # Colores diferentes para cada escenario

# Nmero de preguntas y escenarios
preguntas = range(1, 18)  # Preguntas de 1 a 17
escenarios_labels = list(escenarios.keys())

# Crear una figura con subgrficos (4 filas para las mtricas, 17 columnas para las preguntas)
fig, axs = plt.subplots(len(metricas), len(preguntas), figsize=(20, 12), sharex=True, sharey=True)

# Crear grficos por pregunta y mtrica en cada subcuadrante
for col, pregunta in enumerate(preguntas):
    for row, metrica in enumerate(metricas):
        ax = axs[row, col]
        
        # Recopilar los valores de la mtrica actual para la pregunta actual de todos los escenarios
        valores = [escenarios[escenario][metrica][pregunta - 1] for escenario in escenarios_labels]
        
        # Graficar las barras horizontales
        ax.barh(escenarios_labels, valores, color=colores, alpha=0.8)
        
        # Ajustar los lmites de los ejes
        ax.set_xlim(0, 1.1)

        # Aadir etiquetas solo en los bordes
        if col == 0:
            ax.set_ylabel(metrica)
        if row == len(metricas) - 1:
            ax.set_xlabel(f'Pregunta {pregunta}')
        
        # Hacer la cuadrcula ms tenue
        ax.grid(True, linestyle='--', alpha=0.3)

# Ajustar el espaciado entre los subgrficos
plt.tight_layout()
plt.show()
```
